```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,50)
y=np.sin(x)
def moving_average(values,window_size=3):
  for i in range(len(values)):
   smoothed=[]
  start=max(0,i-window_size//2)
  end=min(len(values),i+window_size//2+1)
  smoothed.append(np.mean(values[start:end]))
  return smoothed
derivative=np.diff(y)
fig,axes=plt.subplots(3,1,figsize=(6,12))
#subplot1-reversed x,y line plot
axes[0].plot(y[::-1],x[::-1])
axes[0].set_title("Reversed x and y plot")
axes[0].set_xlabel("Reversed y")
axes[0].set_ylabel("Reversed x")
#subplot2-smoothed y-values plot
y_smooth=moving_average(y,window_size=5)
axes[1].plot(x,y_smooth)
axes[1].set_title("Smoothed y values")
axes[1].set_xlabel("x")
axes[1].set_ylabel("smoothed y")
# subplot3-Derivative plot
axes[2].plot(x[1::],derivative)
axes[2].set_xlabel("x")
axes[2].set_ylabel("dy/dx")
plt.tight_layout()
plt.show()
```
